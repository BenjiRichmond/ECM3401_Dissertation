```python
import pandas as pd
from langcodes import *

test = pd.read_pickle('users_community_RETWEET_QAT.pkl')
```

```python
def conv_language(lang):
    return Language.make(language=lang).display_name()

test['language'] = test['lang'].apply(conv_language)

bad = ['Unknown language',
       'Unknown language [qme]',
       'Unknown language [qht]',
       'Unknown language [qct]',
       'Unknown language [qst]',
       'Unknown language [qam]']

lang_df = test.loc[~test['language'].isin(bad)]

hash_df = test.explode('hashtags').dropna(subset=['hashtags'])

for comm_id in sorted(test['community'].unique()):
    df_comm_lang = lang_df[lang_df['community'] == comm_id]
    top_langs = (
        df_comm_lang['language']
        .value_counts(normalize=True)
        .mul(100)
        .nlargest(5) 
    )
    
    df_comm_hash = hash_df[hash_df['community'] == comm_id]
    top_hash = (
        df_comm_hash['hashtags']
        .value_counts()
        .nlargest(5)
    )

    print(f"Community {comm_id}:")
    print("  Top 5 Languages:")
    for lang, pct in top_langs.items():
        print(f"    {lang}: {pct:.2f}%")
    
    print("  Top 5 Hashtags:")
    for tag, cnt in top_hash.items():
        print(f"    #{tag}: {cnt}")
    print()


```
